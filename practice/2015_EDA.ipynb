```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\DVK\\Desktop\\Amazon_decade_Sales_project\\practice\\amazon_2015_cleaned.csv")
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
yearly_rev = df.groupby('order_year')['final_amount_inr'].sum().reset_index()
yearly_rev['growth_pct'] = yearly_rev['final_amount_inr'].pct_change() * 100

sns.lineplot(data=yearly_rev, x='order_year', y='final_amount_inr')
plt.title("Yearly Revenue Trend")
plt.show()
```

```python
pivot = df.pivot_table(
    values='final_amount_inr',
    index='order_year',
    columns='order_month',
    aggfunc='sum'
)

sns.heatmap(pivot, cmap='YlGnBu')
plt.title("Monthly Sales Heatmap")
plt.show()
```
